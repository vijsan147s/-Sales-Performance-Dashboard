# 1. Monthly Sales Trend (Line Chart)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sales_Data.csv")

df["Sale_Date"] = pd.to_datetime(df["Sale_Date"], dayfirst=True)
df["Month"] = df["Sale_Date"].dt.month_name()

month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]

monthly_sales = df.groupby("Month")["Sales_Amount"].sum().reindex(month_order)

plt.figure(figsize=(12,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linewidth=3)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# 2. Sales by Region (Bar Chart)

```python
region_sales = df.groupby("Region")["Sales_Amount"].sum().sort_values()

plt.figure(figsize=(8,5))
plt.bar(region_sales.index, region_sales.values)
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

# 3. Product Category Sales (Bar Chart)

```python
category_sales = df.groupby("Product_Category")["Sales_Amount"].sum()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values)
plt.title("Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

# 4. Product Category Distribution (Pie Chart)

```python
category_sales = df.groupby("Product_Category")["Sales_Amount"].sum()

plt.figure(figsize=(7,7))
plt.pie(
    category_sales,
    labels=category_sales.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Sales Distribution by Product Category")
plt.show()
```

# 5. Sales Representative Performance

```python
rep_sales = df.groupby("Sales_Rep")["Sales_Amount"].sum().sort_values()

plt.figure(figsize=(8,5))
plt.barh(rep_sales.index, rep_sales.values)
plt.title("Sales by Sales Representative")
plt.xlabel("Sales Amount")
plt.show()
```

# 6. Customer Type Analysis

```python
customer = df.groupby("Customer_Type")["Sales_Amount"].sum()

plt.figure(figsize=(6,5))
plt.bar(customer.index, customer.values)
plt.title("Customer Type Sales")
plt.ylabel("Sales")
plt.show()
```

# 7. Payment Method Distribution

```python
payment = df.groupby("Payment_Method")["Sales_Amount"].sum()

plt.figure(figsize=(7,7))
plt.pie(
    payment,
    labels=payment.index,
    autopct="%1.1f%%"
)
plt.title("Payment Method Distribution")
plt.show()
```

# 8. Sales Channel Comparison

```python
channel = df.groupby("Sales_Channel")["Sales_Amount"].sum()

plt.figure(figsize=(6,5))
plt.bar(channel.index, channel.values)
plt.title("Sales Channel")
plt.ylabel("Sales")
plt.show()
```

# 9. Discount vs Sales (Scatter Plot)

```python
plt.figure(figsize=(8,5))
plt.scatter(
    df["Discount"],
    df["Sales_Amount"],
    alpha=0.6
)
plt.title("Discount vs Sales")
plt.xlabel("Discount")
plt.ylabel("Sales Amount")
plt.show()
```

# 10. Quantity Sold Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
plt.hist(
    df["Quantity_Sold"],
    bins=10
)
plt.title("Quantity Sold Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

# 11. Sales Amount Distribution

```python
plt.figure(figsize=(8,5))
plt.hist(
    df["Sales_Amount"],
    bins=20
)
plt.title("Sales Amount Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

# 12 Top 10 Highest Sales

```python
top10 = (
    df.groupby("Sales_Rep")["Sales_Amount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(8,5))
plt.bar(top10.index, top10.values)
plt.title("Top Sales Representatives")
plt.ylabel("Sales")
plt.show()
```
